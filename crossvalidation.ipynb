```python
import pandas as pd
```

```python
df = pd.read_csv("/content/load_boston_final_lyst1729240745225.csv")
```

```python
df
```

```python
# training of model:

x = df.iloc[:,1:14]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=78)
```

```python
# scaling so that different scales of the dataset entries becomes equal

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
c_v = cross_val_score(lm,x_train,y_train,scoring="neg_mean_squared_error",cv=10)
```

```python
y_pred = lm.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
lm.score(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```
