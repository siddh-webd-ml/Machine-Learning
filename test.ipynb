# Z test


```python
# Formula for Z test-

# Z = [(mean of the sample-population mean)/(Population sample std/ root(number of samples))]
```

```python
# use for z test:


# 1. whenever we want to find the value of population mean, whem the variance is known.
# 2. Acceptance and rejection region or value of hypothesis
# 3. Whenever the sample is large and variance is known we can say that statistical parameter
# for the population can be known with the help of z test.
# 4. sample size should always be greater than equals to 30
# 5. it only works when the data is SND.
```

```python
# Case Study:

# A passport office claims that they creates the passport within 30 days of filling the application
# Run the hypothesis test to check the claim of the company when the std = 12.5,

# significance level = 0.05,
```

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/passport_lyst1729241000102.csv")
```

```python
df = df.drop("Unnamed: 0",axis=1)
```

```python
df
```

```python
# check for SND:

import seaborn as sns
sns.distplot(df["Processing_time"])
```

```python
import math

from scipy import stats

def z_test(population_mean,Population_std,sample):
  z = (sample.mean()-population_mean)/(Population_std/math.sqrt(len(sample)))
  return z,stats.norm.cdf(z)
```

```python
z_test(30,12.5,df["Processing_time"])
```

```python
# since we can see the value of p is greater than 0.05 which is 0.08 we are going to retain the
# null hypothesis and foing to reject he alternative one.
```
