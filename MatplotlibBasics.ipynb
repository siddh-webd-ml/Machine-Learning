```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,9,4,7]
y=[10,2,5,9,6]

plt.plot(x,y,marker="o",color="g")
```

```python
help(plt.plot)
```

```python
#Bar Plot


x_1=[5,2,9,4,7]
y_1=[10,20,34,55,48]
plt.bar(x_1,y_1,color=["yellow","green","red","blue","pink"])
plt.xlabel("Sales")
plt.ylabel("Profit")

plt.title("Sales Portfolio")
```

```python
help(plt.bar)
```

```python
# Histogram

axis_1=[10,20,3,50]

plt.hist(axis_1)
```

```python
# scatter plot

x=[10,20,30,50]
y=[10,5,8,4]

plt.plot(x,y)
```

```python
plt.scatter(x,y,color="r")
```

```python
# stacked plot


var_1= ["A","B","C","D"]
y1=[10,20,10,45]
y2=[20,25,15,85]
plt.bar(var_1,y2,color="red")
plt.bar(var_1,y1)
```

```python
# Pie Chart

label_1=["Men","Women","others","unidentified"]
y_n=np.array([35,25,25,15])
plt.pie(y_n,labels=label_1,explode=[0.1,0.2,0.3,0.1])
plt.legend()
```
