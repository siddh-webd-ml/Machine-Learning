#Multivariate Analysis

```python
#data cleaning---> Duplicate values and we are going to treat null values.
# feature engineering---> creation of features based on existing ones.
# data analysis--->performing analysis on data based on different factors.
```

```python
#Duplicate Values


df.duplicated().sum()
```

```python
df=df.drop_duplicates()
df
```

```python
#null values
```

```python
  df.info()
```

```python
df.isnull().sum()
```

```python
df["Gender"].fillna(df["Gender"].mode()[0])
```

```python
df["Gender"]=df["Gender"].mode()[0]
```

```python
df["LoanAmount"] = pd.to_numeric(df["LoanAmount"], errors="coerce")
```

```python
df["LoanAmount"]=df["LoanAmount"].fillna(df["LoanAmount"].median)
```

```python
#feature engineering

df
```

```python
df["LoanAmount"] = pd.to_numeric(df["LoanAmount"], errors="coerce")
```

```python
df["Ratio"]=df["ApplicantIncome"]/df["LoanAmount"]
```

```python
df
```

```python
df["Ratio"].describe()
```

```python
#binning
df
#if person is married,,gradduate and not self employed---> stable
#else:
#    unstable
```

```python
def bin(x):
  if (x["Married"]=="Yes") & (x["Education"]=="Graduate") & (x["Self_Employed"]=="No"):
    return "Stable"
  else:
    return "UnStable"
```

```python
df["Bin"]=df.apply(bin,axis=1)
```

#Group BY

```python
#give the overall loanamount for the graduate people

df.groupby("Education").agg({"LoanAmount":"sum"})
```

```python
df.groupby("Gender").agg({"ApplicantIncome": "sum"})
```

```python
#What is the max loan amount asked by the male gender,

df.groupby("Gender").agg({"LoanAmount":"max"})
```
