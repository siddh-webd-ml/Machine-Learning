```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/loan_data_set.csv")
```

```python
df
```

```python
#univariate Categorical
# x=df["Education"].value_counts()
sns.catplot(x="Education",data=df,kind="count")
```

```python
# univariate numerical feature:

sns.distplot(df["ApplicantIncome"],hist=False,color="red")
```

```python
plt.figure(figsize=(15,5))
sns.histplot(df["ApplicantIncome"])
```

```python
#bIVARIATE NUMERICAL :

sns.jointplot(x="ApplicantIncome",y="LoanAmount",data=df,kind="scatter",hue="Loan_Status")
```

```python
  sns.pairplot(data=df,hue="Loan_Status")
```

```python
#bivariate categorical

sns.boxplot(x="Gender",y="LoanAmount",data=df)
```

```python
sns.catplot(x="Gender",y="LoanAmount",data=df,kind="swarm ")
```

```python
sns.catplot(x="Gender",y="LoanAmount",data=df,kind="violin")
```
